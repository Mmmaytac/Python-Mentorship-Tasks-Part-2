### Task 1: Visualize the distribution of the price column using a histogram.


```python
plt.hist(df["price"],bins=70)
plt.xlabel("price")
plt.ylabel("frequency")
plt.show()
```

### Task 2: Show the relationship between carat and price using a scatter plot.



### Task 3: Visualize the average price by cut category using a bar chart.



```python
avg_price=df.groupby("cut")["price"].mean()
avg_price.plot(kind="bar")
plt.ylabel("avg price")
plt.xlabel("cut")
plt.show()
```

### Task4 : Show the distribution of price by color using a box plot.



```python
sns.boxplot(x="color",y="price", data=df)
plt.show()
```

### Task 5: Display the number of diamonds for each clarity category.



```python
df["clarity"].value_counts().plot(kind="bar")
plt.show()
```

### Task 6: Create side-by-side histograms for depth and table.



```python
plt.subplot(1,2,1)
plt.hist(df["depth"], bins=30)

plt.subplot(1,2,2)
plt.hist(df["table"], bins=30)

plt.show
```

### Task 7: Visualize the relationship between x, y, and z using a pair plot.



```python
sns.pairplot(df[['x','y','z']])
plt.show()
```

### Task 8: Create a correlation heatmap between price and numerical variables.

```python
corr=df[["carat","depth","table","price","x","y","z"]].corr()
corr
```

```python
sns.heatmap(corr, cmap="coolwarm", annot=True)
plt.show()
```
